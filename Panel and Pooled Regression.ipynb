## Panel data Regression

```python
import pandas as pd
from linearmodels.panel import PanelOLS
```

```python
# Load dataset
df_panel = pd.read_excel(r"C:\Users\HP\Downloads\SouthAsianPanelData.xlsx")
df_panel
```

```python
# create a multi_index dataFrame for panel data
df_panel.set_index(['Country', 'Year'], inplace = True)
```

```python
df_panel
```

```python
# Independent and dependent feature
x = df_panel[['Imports','Remittances','Exports','Consumption','Government Consumption']]
y = df_panel['GDP']
```

```python
# create a panel regression model
panel_model = PanelOLS(y, x, entity_effects = True, time_effects = True)
```

```python
# fit the model
panel_result = panel_model.fit()
```

```python
panel_result
```

```python
panel_result.std_errors
```

```python
# Prediction
panel_result.predict()
```

## Pooled Regression

```python
from linearmodels.panel import PooledOLS
```

```python
df = pd.read_excel(r"C:\Users\HP\Downloads\SouthAsianPanelData.xlsx")
df
```

```python
df_pooled = df[df['Year'] == 2021]
```

```python
df_pooled.set_index(['Country','Year'], inplace = True)
```

```python
df_pooled
```

```python
df_pooled.shape
```

```python
# Independent and Dependent features
x_pooled = df_pooled[['Imports','Remittances','Exports','Consumption','Government Consumption']]
y_pooled = df_pooled['GDP']
```

```python
# Create a pooled regression model
pooled_model = PooledOLS(y_pooled, x_pooled)
```

```python
# model fit
pooled_result = pooled_model.fit()
```

```python
pooled_result
```

```python
pooled_result.predict()
```
